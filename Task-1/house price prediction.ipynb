```python
#Droping the man entry values
df.dropna(axis=0,inplace=True)
```

```python
df.info()
```

```python
df.ocean_proximity.value_counts()
```

```python
df=df.join(pd.get_dummies(df['ocean_proximity'])).drop(['ocean_proximity'],axis=1)
```

```python
## Feature Engineering 
df['bedroom_ratio']=df['total_bedrooms']/df['total_rooms']
df['household_rooms']=df['total_rooms']/df['households']
df
```

```python
plt.figure(figsize=(16,8))
sbs.heatmap(df.corr(),annot=True ,cmap="YlGnBu")
```

```python
## Modal preparation
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X = df.drop('median_house_value', axis = 1)
y = df['median_house_value']
LR=LinearRegression()
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
LR.fit(x_train,y_train)
predict = LR.predict(x_test)
```

```python
#prediction value and actual value
print('Predicted Value :',predict[0])
print('Actual Value :',y_test.values[0])
```

```python
#score
LR.score (x_test, y_test)
```
